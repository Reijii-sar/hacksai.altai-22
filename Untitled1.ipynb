```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from catboost import CatBoostClassifier, Pool
from sklearn.utils import shuffle, compute_class_weight
import warnings
import time
from IPython.display import display_html
from itertools import chain,cycle
```

```python
train = pd.read_csv('train_dataset_train.csv')
test = pd.read_csv('test_dataset_test.csv')
submission = pd.read_csv('sample_submission.csv')

id_language = pd.read_csv('id_language.csv')
id_country = pd.read_csv('id_country.csv')
```

```python
train.dropna()
```

```python
train=train.drop(columns=['Дата_Рождения'])
test=test.drop(columns=['Дата_Рождения'])
train.dtypes[train.dtypes == "object"].index.values
mass_object = train.dtypes[train.dtypes == "object"].index.values
#mass_object = np.append(mass_object, "Статус")
train = train.drop(mass_object, axis = 1)
test = test.drop(mass_object, axis = 1)
```

```python
#корректируем тип данных для корректной отработки модели
def type_cor (data, column):
    for i in column:
        data[i]=data[i].astype('int')
    return data
```

```python
train = train.fillna(0)
test = test.fillna(0)

column = [
    'ID',
    'Код_группы',
    'Год_Поступления',
    'Год_Окончания_УЗ',
    'Пособие',
    'Общежитие',
    'Наличие_Матери',
    'Наличие_Отца',
    'Опекунство',
    'Село',
    'Иностранец',
    'КодФакультета',
    'СрБаллАттестата',
]
train['Статус']=train['Статус'].astype('int')
#train_2['Пособие']=train_2['Пособие'].astype('int')
#train_2['Наличие_Отца']=train_2['Наличие_Отца'].astype('int')
#train_2['Опекунство']=train_2['Опекунство'].astype('int')

#test_2['Пособие']=test_2['Пособие'].astype('int')
#test_2['Наличие_Отца']=test_2['Наличие_Отца'].astype('int')
#test_2['Опекунство']=test_2['Опекунство'].astype('int')
train = type_cor (train, column)
test = type_cor (test, column)
```

```python
train.info()
```

```python
X = train.drop(columns=['Статус'])
y = train['Статус']
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25, random_state=12345)
```

```python
#upsampe
def up (features, target):
    a = features[target == -1]
    b = features[target == 3]
    c = features[target == 4] 
    a_1 = target[target == -1]
    b_1 = target[target == 3]
    c_1 = target[target == 4]
    f_up = pd.concat ([a]*80+[b]*10+[c]*8)
    t_up = pd.concat ([a_1]*80+[b_1]*10+[c_1]*8)
    f_up, t_up = shuffle(f_up, t_up, random_state=12345)
    return f_up, t_up
X_train, y_train = up (X_train, y_train)
```

```python
column_category = [
   'ID',
   'Код_группы',
#    'Год_Поступления',
#    'Пол',
#    'Основания',
#    'Изучаемый_Язык',
#    'Дата_Рождения',
#    'Уч_Заведение',
#    'Где_Находится_УЗ',
#    'Год_Окончания_УЗ',
    'Пособие',
#    'Страна_ПП',
#    'Регион_ПП',
#    'Город_ПП',
    'Общежитие',
    'Наличие_Матери',
    'Наличие_Отца',
#    'Страна_Родители',
    'Село',
    'Иностранец',
    'КодФакультета',
#    'СрБаллАттестата',
#    'Семья',
#    'Ур_образования',
] 
pool_train = Pool(X_train, y_train, cat_features = column_category)
pool_valid = Pool(X_valid, cat_features = column_category)
pool_test_solution = Pool(test, cat_features = column_category)
```

```python
classes = np.unique(y_train)
weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train) 
class_weights = dict(zip(classes, weights))
```

```python
model = CatBoostClassifier(
    task_type='GPU',
    random_state = 12345,
    iterations = 3000,
    learning_rate = 0.1,
    depth = 12,
    loss_function= 'MultiClass',
    eval_metric='TotalF1',
    grow_policy = 'Depthwise',
    min_data_in_leaf = 3000,
#    class_weights = class_weights,
##    average = 'Macro',
    bootstrap_type ='Bernoulli',
##    subsample = 0.8,
##    leaf_estimation_method = 'Newton',
    verbose = 10,
#    classes_count = 3,
    ignored_features = ['ID']#, 'Код_группы']#'Год_Поступления', 'Год_Окончания_УЗ', 'Между_поступлениями', 'Общежитие','Село',
#                        'Изучаемый_Язык','Иностранец','Семья']
    )
model.fit(pool_train, early_stopping_rounds=500, eval_set=(X_valid, y_valid), plot=True, use_best_model=True)
```

```python
pred_valid = model.predict(pool_valid)
```

```python
model.get_feature_importance(prettified=True)
```

```python
print(classification_report(y_valid, pred_valid))
```

```python
pred_solution = model.predict(pool_test_solution)
```

```python
submission["Статус"] = pred_solution
```
